**Q11: Verify put-call parity for European options**
- C + Ke^(-rT) = P + S0
- C = European call price
- P = European put price
- K = Strike price
- r = Risk-free rate
- T = Time to maturity
- S_o = Initial stock price

```python
import numpy as np
import matplotlib.pyplot as plt

# Given values
S0 = 100  # Initial stock price
K = 100  # Strike price
r = 0.05  # Risk-free rate
T = 0.75  # Time to maturity in years
C = 8.77  # European call price
P = 5.08  # European put price

# Calculate Ke^(-rT)
Ke_rt = K * np.exp(-r * T)

# Verify put-call parity
lhs = C + Ke_rt
rhs = P + S0

print(f"Left-hand side: {lhs:.2f}")
print(f"Right-hand side: {rhs:.2f}")

if np.abs(lhs - rhs) < 0.1:  # allowing for small rounding errors
    print("Put-call parity holds approximately.")
else:
    print("Put-call parity does not hold.")

# Graph the put-call parity relationship
S = np.linspace(80, 120, 100)  # stock prices
C_values = np.maximum(S - K, 0)  # call option payoff
P_values = np.maximum(K - S, 0)  # put option payoff

plt.figure(figsize=(10, 6))
plt.plot(S, C_values + Ke_rt, label='C + Ke^(-rT)')
plt.plot(S, P_values + S, label='P + S')
plt.xlabel('Stock Price')
plt.ylabel('Value')
plt.title('Put-Call Parity Relationship')
plt.legend()
plt.grid(True)
plt.show()
```

The put-call parity holds approximately, with a small difference likely due to rounding errors.

This confirms that the European call and put options satisfy put-call parity, which is a fundamental relationship in option pricing theory.

**Using Binomial Tree Model** to price European and American put options (similar to Team A and Team B's work):

```python
import numpy as np

# Given parameters
S0 = 100  # Initial stock price
K = 100  # Strike price
r = 0.05  # Risk-free rate
T = 0.75  # Time to maturity in years
sigma = 0.20  # Volatility
N = 250  # Number of steps (used by Team A)

# Calculate delta t and other parameters
dt = T / N
u = np.exp(sigma * np.sqrt(dt))
d = 1 / u
p = (np.exp(r * dt) - d) / (u - d)

# Initialize arrays to store option prices
P_european = np.zeros(N + 1)
P_american = np.zeros(N + 1)

# Calculate option prices at maturity
for i in range(N + 1):
    S_T = S0 * (u ** (N - i)) * (d ** i)
    P_european[i] = np.maximum(K - S_T, 0)
    P_american[i] = np.maximum(K - S_T, 0)

# Backward induction to calculate option prices
for i in range(N - 1, -1, -1):
    for j in range(i + 1):
        S_t = S0 * (u ** (i - j)) * (d ** j)
        P_european[j] = np.exp(-r * dt) * (p * P_european[j] + (1 - p) * P_european[j + 1])
        P_american[j] = np.maximum(K - S_t, np.exp(-r * dt) * (p * P_american[j] + (1 - p) * P_american[j + 1]))

# Round the final prices to two decimal places
P_european_rounded = round(P_european[0], 2)
P_american_rounded = round(P_american[0], 2)

# Print the results
print(f"European put price: {P_european_rounded:.2f}")
print(f"American put price: {P_american_rounded:.2f}")

if P_european_rounded <= P_american_rounded:
    print("European put ≤ American put confirmed.")
else:
    print("European put > American put. This is not expected.")
```

**Q19 : European Call and Put Prices vs. Stock Prices**




The European call and put prices are calculated using the binomial tree model with the given parameters:
- S0 = 100
- K = 100
- r = 0.05
- T = 0.75
- sigma = 0.20
- N = 250 (number of steps)

Calculating the European call and put prices for different stock prices (S0 = 80, 90, 100, 110, 120) and graphing the results -

The European call and put prices are given by the binomial tree model:
- $C = e^(-rT) * E[max(S_T - K, 0)]$
- $P = e^(-rT) * E[max(K - S_T, 0)]$

The call price increases with the stock price, while the put price decreases with the stock price.


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Given parameters
K = 100  # Strike price
r = 0.05  # Risk-free rate
T = 0.75  # Time to maturity in years
sigma = 0.20  # Volatility
N = 250  # Number of steps

# Stock prices
S0_values = np.array([80, 90, 100, 110, 120])

# Initialize arrays to store option prices
C_european = np.zeros(len(S0_values))
P_european = np.zeros(len(S0_values))

# Calculate European call and put prices for different stock prices
for i, S0 in enumerate(S0_values):
    # Calculate delta t and other parameters
    dt = T / N
    u = np.exp(sigma * np.sqrt(dt))
    d = 1 / u
    p = (np.exp(r * dt) - d) / (u - d)

    # Initialize arrays to store option prices
    C = np.zeros(N + 1)
    P = np.zeros(N + 1)

    # Calculate option prices at maturity
    for j in range(N + 1):
        S_T = S0 * (u ** (N - j)) * (d ** j)
        C[j] = np.maximum(S_T - K, 0)
        P[j] = np.maximum(K - S_T, 0)

    # Backward induction to calculate option prices
    for j in range(N - 1, -1, -1):
        for k in range(j + 1):
            C[k] = np.exp(-r * dt) * (p * C[k] + (1 - p) * C[k + 1])
            P[k] = np.exp(-r * dt) * (p * P[k] + (1 - p) * P[k + 1])

    C_european[i] = round(C[0], 2)
    P_european[i] = round(P[0], 2)

# Create a DataFrame for European option prices
european_prices_df = pd.DataFrame({
    'Stock Price': S0_values,
    'European Call': C_european,
    'European Put': P_european
})
print("European Call and Put Prices vs. Stock Prices:")
display(european_prices_df)

# Graph the results
plt.figure(figsize=(10, 6))
plt.plot(S0_values, C_european, label='European Call')
plt.plot(S0_values, P_european, label='European Put')
plt.xlabel('Stock Price')
plt.ylabel('Option Price')
plt.title('European Call and Put Prices vs. Stock Prices')
plt.legend()
plt.grid(True)
plt.show()
```

The graph shows the European call and put prices against stock prices (S0 = 80, 90, 100, 110, 120). The call price increases with the stock price, while the put price decreases with the stock price.

The computation shows that the European call option price increases as the underlying stock price increases, while the European put option price decreases as the stock price increases. This is consistent with financial intuition, as a call option becomes more valuable when the stock price rises, and a put option becomes less valuable.

The graph provides a visual representation of this relationship, illustrating the non-linear payoff structure of options.
